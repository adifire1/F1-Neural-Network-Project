```python
import matplotlib.pyplot as plt
import fastf1.plotting


fastf1.Cache.enable_cache('../doc_cache')  # replace with your cache directory

# enable some matplotlib patches for plotting timedelta values and load
# FastF1's default color scheme
fastf1.plotting.setup_mpl()

# load a session and its telemetry data
ultraSoftLength = []
hyperSoftLength = []
superSoftLength = []
softLength = []
mediumLength = []
hardLength = []
superHardLength = []
for j in range(1,21):
    session = fastf1.get_session(2018, j, 'Race')
    session.load()
    mag_laps = session.laps.pick_driver('MAG')
    for i in range(len(mag_laps)):
        if(mag_laps.iloc[i-1]['TyreLife']>mag_laps.iloc[i]['TyreLife'] or i == len(mag_laps)-1):
            compound = mag_laps.iloc[i-1]['Compound']
            if(compound == 'ULTRASOFT'):
                ultraSoftLength.append(mag_laps.iloc[i-1]['TyreLife'])
            if(compound == 'HYPERSOFT'):
                hyperSoftLength.append(mag_laps.iloc[i-1]['TyreLife'])
            if(compound == 'SUPERSOFT'):
                superSoftLength.append(mag_laps.iloc[i-1]['TyreLife'])
            if(compound == 'SOFT'):
                softLength.append(mag_laps.iloc[i-1]['TyreLife'])
            if(compound == 'MEDIUM'):
                mediumLength.append(mag_laps.iloc[i-1]['TyreLife'])
            if(compound == 'HARD'):
                hardLength.append(mag_laps.iloc[i-1]['TyreLife'])
            if(compound == 'SUPERHARD'):
                superHardLength.append(mag_laps.iloc[i-1]['TyreLife'])
                
```

```python
print(ultraSoftLength,supersoftLength, softLength, mediumLength, hardLength)
```
